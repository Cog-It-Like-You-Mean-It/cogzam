```python
# Example data structure
import pickle
from collections import Counter


class Database:
    database = {}

    def __init__(self):
        pass

    def store_fingerprints(self, fingerprints, song_title, times):
        # key: (freq, neighbor_freq, time_delta)), value: (song, time_of_occurence)
        # store fingerprints of 1 song

        for i in range(len(fingerprints)):
            f_peak = fingerprints[i]  # group of fingerprints by the same peak

            for f in f_peak:
                if f in self.database.keys():
                    self.database[f].append((song_title, times[i]))
                else:
                    self.database[f] = [(song_title, times[i])]

    def search_song(self, fingerprints, times):
        # key: fingerprints, value: (song, time_of_occurance)
        # all fingerprints of a song are given, return most likely song

        # find the time_of_occurance for first fingerprint
        l = []
        count = 0
        for f_peak in fingerprints:
            # print(f_peak)
            for f in f_peak:

                if f in self.database.keys():
                    for match in self.database[f]:
                        l.append((match[0], match[1] - times[count]))

            count += 1

            # consider returning counts here or in notebook
        c = Counter(l)
        return c

    def load_database(self, file_name):
        import pickle

        with open("fingerprints.pkl", mode="rb") as fingerprints_file:
            self.database = pickle.load(fingerprints_file)

        return

    def save_database(self):
        with open("fingerprints.pkl", mode="wb") as fingerprints_file:
            pickle.dump(self.database, fingerprints_file)
```

```python
sg.theme('Python')

layoutStart = [[sg.Button('Create New Database'), sg.Button('Load Database')],
          [sg.Button('Close')]]
windowStart = sg.Window('Cog*Zam: Create or Load Database', layoutStart)

database = Database()


while True:
    event, values = windowStart.read()      
    if event == sg.WIN_CLOSED or event == 'Close':
        break
    if event == 'Create New Database':
        raise NotImplementedError
        break
    if event == 'Load Database':
        layoutLoad = [[sg.Text('Select File'), sg.InputText('Database.pkl'), sg.FileBrowse()],
                      [sg.Button('OK'), sg.Button('Cancel')]]
        windowLoad = sg.Window('Cog*Zam: Load Database', layoutLoad)
        
        while True:
            event, values = windowLoad.read()
            if event == sg.WIN_CLOSED or event == 'Cancel':
                break
            if event == 'OK':
                database.load_database()
                break
        windowLoad.close()
        break
windowStart.close()
if event != sg.WIN_CLOSED and event != 'Close':
    layoutHome = [[sg.Button('Add Song'), sg.Button('Search for Song')],
                  [sg.Button('Close')]]
    windowHome = sg.Window('Cog*Zam: Function Select', layoutHome)

    while True:
        event, values = windowHome.read()
        if event == sg.WIN_CLOSED or event == 'Close':
            break
        if event == 'Add Song':
            layoutAdd = [[sg.Text('Select File'), sg.InputText('Song.mp3'), sg.FileBrowse()],
                         [sg.Text('Input Song Name'), sg.InputText('SongName')],
                         [sg.Button('OK'), sg.Button('Cancel')]]
            windowAdd = sg.Window('Cog*Zam: Add Song', layoutAdd)

            while True:
                event, values = windowAdd.read()
                if event == sg.WIN_CLOSED or event == 'Cancel':
                    break
                if event == 'OK':
                    cz.import_song(database, values[0], values[1])
                    break
            windowAdd.close()
        if event == 'Search for Song':
            layoutSearch = [[sg.Text('Play Song')]]
            windowSearch = sg.Window('Cog*Zam: Song Search', layoutSearch)
            answer = cz.microphone_match(database)
            windowSearch.close()

            layoutAnswer = [[sg.Text('Song is: ' + answer)],
                            [sg.Button('OK')]]
            windowAnswer = sg.Window('Cog*Zam: Output', layoutAnswer)
            
            while True:
                event, values = windowAnswer.read()
                if event == sg.WIN_CLOSED or event == 'OK':
                    break
            windowAnswer.close()
    windowHome.close()

```
